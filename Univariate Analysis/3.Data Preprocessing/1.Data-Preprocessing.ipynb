#NOTE: Ways to Fill Empty, null or NaN Values.
For Quantitative Columns (Numerical Data):
1. Replave the missing Value with Central Tendency - Mean Median Mode
2. Entire Row Delete
3. Respective to the problem statement, have to fill the missing values.
4. Consider this as a Semi-Supervised Learning,   Create a model, so that we can predict the missin Values.
   For example, in this dataset, take salary column as Output and consider remaining as Input.

For Qualitative Columns (Categorical Data):
1. Replave the missing Value with Central Tendency - Mode.
2. 4. Consider this as a Semi-Supervised Learning,   Create a model, so that we can predict the missin Values.

```python
dataset.fillna(dataset["salary"].mean(), inplace=True) # Just for learning, replacing mean for this dataset does not suit, becuase Not Placed Candidate should get salary. 
dataset
```

```python
#To find mean (for numerical data) - seperate qulantitative and qualitative data from dataset
def quanQual(dataset):
    quan = []
    qual = []
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan, qual 
```

```python
quan, qual = quanQual(dataset)
quan
```

```python
dataset[quan]
```

```python
#Note this part is for learning Purpose - mean does not suit for this dataset, becuase Not Placed Candidate should not get any salary.
#To Fill missing values of Multiple Columns use - IMPUTER
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df = imp.transform(dataset[quan])
```

```python
df = pd.DataFrame(df, columns = quan) #where df - data (Existing Data), columns = quan - For ColumName
df
```
